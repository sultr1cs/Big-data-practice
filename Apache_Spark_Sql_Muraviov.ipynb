```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Бібліотеки імпортовано успішно!")
```

```python
try:
    df = pd.read_csv('results.csv')
    print(f"Дані завантажено успішно!")
    print(f"Розмір датасету: {df.shape}")
    print(f"\nПерші 5 рядків:")
    print(df.head())
    print(f"\nІнформація про стовпці:")
    print(df.info())
except FileNotFoundError:
    print("ПОМИЛКА: Файл results.csv не знайдено!")
    print("Будь ласка, завантажте файл у Colab перед виконанням цього блоку")
```

```python
print("=" * 60)
print("ПИТАННЯ 1: Скільки матчів зіграно командами")
print("=" * 60)

# Німеччина
germany_home = df[df['home_team'] == 'Germany'].shape[0]
germany_away = df[df['away_team'] == 'Germany'].shape[0]
germany_total = germany_home + germany_away

print(f"\nНімеччина:")
print(f"  Матчів вдома: {germany_home}")
print(f"  Матчів на виїзді: {germany_away}")
print(f"  Всього матчів: {germany_total}")

# Україна
ukraine_home = df[df['home_team'] == 'Ukraine'].shape[0]
ukraine_away = df[df['away_team'] == 'Ukraine'].shape[0]
ukraine_total = ukraine_home + ukraine_away

print(f"\nУкраїна:")
print(f"  Матчів вдома: {ukraine_home}")
print(f"  Матчів на виїзді: {ukraine_away}")
print(f"  Всього матчів: {ukraine_total}")
```

```python
print("\n" + "=" * 60)
print("ПИТАННЯ 2: Скільки матчів зіграно в містах")
print("=" * 60)

# Київ
kyiv_matches = df[df['city'] == 'Kyiv'].shape[0]
print(f"\nКиїв: {kyiv_matches} матчів")

# Лондон
london_matches = df[df['city'] == 'London'].shape[0]
print(f"Лондон: {london_matches} матчів")
```

```python
print("\n" + "=" * 60)
print("ПИТАННЯ 3: Топ-10 команд за голами вдома")
print("=" * 60)

# Групуємо по домашній команді та сумуємо голи
home_goals = df.groupby('home_team')['home_score'].sum().sort_values(ascending=False)
top10_home = home_goals.head(10)

print("\nТоп-10 команд за кількістю голів вдома:")
for i, (team, goals) in enumerate(top10_home.items(), 1):
    print(f"{i:2d}. {team:20s} - {int(goals):5d} голів")
```

```python
print("\n" + "=" * 60)
print("ПИТАННЯ 4: Топ-10 команд за голами за весь час")
print("=" * 60)

# Голи вдома
home_total = df.groupby('home_team')['home_score'].sum()

# Голи на виїзді
away_total = df.groupby('away_team')['away_score'].sum()

# Об'єднуємо та сумуємо всі голи
all_teams = set(home_total.index) | set(away_total.index)
total_goals = {}

for team in all_teams:
    home = home_total.get(team, 0)
    away = away_total.get(team, 0)
    total_goals[team] = home + away

# Сортуємо та беремо топ-10
total_goals_sorted = pd.Series(total_goals).sort_values(ascending=False)
top10_total = total_goals_sorted.head(10)

print("\nТоп-10 команд за загальною кількістю голів:")
for i, (team, goals) in enumerate(top10_total.items(), 1):
    print(f"{i:2d}. {team:20s} - {int(goals):5d} голів")
```

```python
print("\n" + "=" * 60)
print("ПИТАННЯ 5: Середній бал команд (Топ-10)")
print("=" * 60)

# Середній бал вдома
home_avg = df.groupby('home_team')['home_score'].mean()
home_count = df.groupby('home_team').size()

# Середній бал на виїзді
away_avg = df.groupby('away_team')['away_score'].mean()
away_count = df.groupby('away_team').size()

# Обчислюємо зважений середній бал
all_teams = set(home_avg.index) | set(away_avg.index)
avg_scores = {}

for team in all_teams:
    h_avg = home_avg.get(team, 0)
    h_cnt = home_count.get(team, 0)
    a_avg = away_avg.get(team, 0)
    a_cnt = away_count.get(team, 0)

    total_goals = h_avg * h_cnt + a_avg * a_cnt
    total_matches = h_cnt + a_cnt

    if total_matches > 0:
        avg_scores[team] = total_goals / total_matches

# Сортуємо та беремо топ-10
avg_scores_sorted = pd.Series(avg_scores).sort_values(ascending=False)
top10_avg = avg_scores_sorted.head(10)

print("\nТоп-10 команд за середнім балом за матч:")
for i, (team, avg) in enumerate(top10_avg.items(), 1):
    print(f"{i:2d}. {team:20s} - {avg:.2f} голів/матч")
```

```python
print("\n" + "=" * 60)
print("ЗАВДАННЯ 2: Операції JOIN")
print("=" * 60)

# DataFrame 1: Employees
employees_data = {
    'employee_id': [1, 2, 3, 4, 5, 6],
    'name': ['John', 'Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'department': ['IT', 'HR', 'IT', 'Sales', 'IT', 'HR']
}
employees_df = pd.DataFrame(employees_data)

print("\nEmployees DataFrame:")
print(employees_df)

# DataFrame 2: Projects
projects_data = {
    'project_id': [101, 102, 103, 104, 105],
    'project_name': ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon'],
    'employee_id': [1, 2, 3, None, 1]
}
projects_df = pd.DataFrame(projects_data)

print("\nProjects DataFrame:")
print(projects_df)
```

```python
print("\n" + "=" * 60)
print("1. INNER JOIN")
print("=" * 60)

inner_join = pd.merge(employees_df, projects_df, on='employee_id', how='inner')
print("\nВсі працівники, залучені до проектів:")
print(inner_join)

# Фільтр для відділу IT
print("\n" + "-" * 60)
print("Працівники з відділу IT, залучені до проектів:")
print("-" * 60)
inner_join_it = inner_join[inner_join['department'] == 'IT']
print(inner_join_it)

# Збереження результатів
inner_join.to_parquet('inner_join.parquet', index=False)
inner_join_it.to_parquet('inner_join_it.parquet', index=False)
print("\n✓ Результати збережено у inner_join.parquet та inner_join_it.parquet")
```

```python
print("\n" + "=" * 60)
print("2. LEFT JOIN")
print("=" * 60)

left_join = pd.merge(employees_df, projects_df, on='employee_id', how='left')
print("\nВсі працівники та їх проекти (якщо є):")
print(left_join)

# Агрегування: кількість працівників у кожному відділі
print("\n" + "-" * 60)
print("Кількість працівників у кожному відділі:")
print("-" * 60)
dept_count = employees_df.groupby('department').size().reset_index(name='employee_count')
print(dept_count)

# Збереження результатів
left_join.to_parquet('left_join.parquet', index=False)
print("\n✓ Результати збережено у left_join.parquet")
```

```python
print("\n" + "=" * 60)
print("3. RIGHT JOIN")
print("=" * 60)

right_join = pd.merge(employees_df, projects_df, on='employee_id', how='right')
print("\nВсі проекти та працівники, що до них залучені:")
print(right_join)

# Збереження результатів
right_join.to_parquet('right_join.parquet', index=False)
print("\n✓ Результати збережено у right_join.parquet")
```

```python
print("\n" + "=" * 60)
print("4. FULL OUTER JOIN")
print("=" * 60)

full_outer_join = pd.merge(employees_df, projects_df, on='employee_id', how='outer')
print("\nВсі працівники та проекти (повне об'єднання):")
print(full_outer_join)

# Збереження результатів
full_outer_join.to_parquet('full_outer_join.parquet', index=False)
print("\n✓ Результати збережено у full_outer_join.parquet")
```

```python
print("\n" + "=" * 60)
print("5. CROSS JOIN")
print("=" * 60)

# Cross join (декартовий добуток)
employees_df['key'] = 1
projects_df['key'] = 1
cross_join = pd.merge(employees_df, projects_df, on='key', how='outer').drop('key', axis=1)

print("\nДекартовий добуток працівників та проектів:")
print(f"Всього комбінацій: {len(cross_join)}")
print(cross_join.head(15))  # Показуємо перші 15 рядків

# Збереження результатів
cross_join.to_parquet('cross_join.parquet', index=False)
print("\n✓ Результати збережено у cross_join.parquet")
```
